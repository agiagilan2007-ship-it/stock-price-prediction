# Stock price prediction: LSTM + Transformer (multi-step)
This notebook runs a compact end-to-end example for AAPL (60-step window -> 7-day horizon).

```python
import sys
sys.path.append('src')
from data_loader import download_ohlcv, create_features, create_windows, train_val_test_split
from model import build_lstm_model, build_transformer_model
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = download_ohlcv('AAPL')
df = create_features(df)
feature_cols = ['Open','High','Low','Close','Volume']
X, y, meta = create_windows(df, window=60, horizon=7, feature_cols=feature_cols)
X_train, y_train, X_val, y_val, X_test, y_test = train_val_test_split(X, y)
print('shapes', X_train.shape, y_train.shape)
```

```python
model = build_lstm_model(input_shape=X_train.shape[1:], horizon=7)
hist = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, batch_size=32)
```

```python
pred = model.predict(X_test[:10])
print('pred shape', pred.shape)
```

You can run training with `python src/train.py --model lstm --ticker AAPL` and evaluate with `python src/evaluate.py --model models/..../best.h5`.
